```python
with open("gears.txt", "r") as f:
    text = f.read().split("\n\n")

gears = []

for gear in text:
    new_gear = []
    new_gear_rows = gear.replace("blank", "__").strip(" ").split("\n")
    for i in range(4 - len(new_gear_rows)):
        new_gear.append(["__"] * 12)
    for some_gear_row in new_gear_rows:
        some_gear_row = some_gear_row.strip("\n").split(", ")
        if some_gear_row != ['']:
            new_gear.append(some_gear_row)
    gears.append(new_gear)
```

```python
for gear in gears:
    for gear_row in gear:
        print('\t'.join(gear_row))
    print("\n" + "_" * 90 + "\n")
```

```python
# Python function to generate cyclic rotations of a list

def cyclic_rotations(lst):
    rotations = []
    n = len(lst)
    for i in range(n):
        # Rotate the list by slicing and appending
        rotated = lst[i:] + lst[:i]
        rotations.append(rotated)
    return rotations

# Example usage
input_list = [1, 2, 3]
output_rotations = cyclic_rotations(input_list)

output_rotations
```
